```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Data points for age and salary
age = np.array([25, 26, 27, 30, 35, 37]).reshape(-1, 1)
salary = np.array([25, 26, 29, 40, 75, 79])

# Create a linear regression model and fit the data
model = LinearRegression()
model.fit(age, salary)

# Predict salary values using the model
age_range = np.linspace(24, 38, 100).reshape(-1, 1)  # Extended range for smooth line
predicted_salary = model.predict(age_range)

# Plotting the data points and the linear regression line
plt.scatter(age, salary, color='blue', label='Data Points')
plt.plot(age_range, predicted_salary, color='red', label='Linear Regression')

# Adding labels and title
plt.xlabel('Age (Years)')
plt.ylabel('Salary (in Thousands)')
plt.title('Linear Regression: Age vs Salary')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Predict salary for age 40
predicted_salary_at_40 = model.predict([[40]])
predicted_salary_at_40
```

**the growth of the function**

```python
slope = model.coef_
intercept = model.intercept_
print(f"Slope: {slope}, Intercept: {intercept}")
```

```python
age_range_wide = np.linspace(20, 50, 100).reshape(-1, 1)
predicted_salary_wide = model.predict(age_range_wide)

plt.plot(age_range_wide, predicted_salary_wide, color='red', label='Linear Regression Growth')
plt.scatter(age, salary, color='blue', label='Data Points')
plt.xlabel('Age (Years)')
plt.ylabel('Salary (in Thousands)')
plt.title('Growth of Function: Age vs Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Data points (age, salary)
age = np.array([25, 26, 27, 30, 35, 37]).reshape(-1, 1)
salary = np.array([25, 26, 29, 40, 75, 79])

# Transform the features to polynomial (degree 2 for quadratic curve)
poly = PolynomialFeatures(degree=2)  # You can increase the degree for more complex curves
age_poly = poly.fit_transform(age)

# Fit the model
model = LinearRegression()
model.fit(age_poly, salary)

# Predict over a wider age range
age_range_wide = np.linspace(20, 50, 100).reshape(-1, 1)
age_range_wide_poly = poly.transform(age_range_wide)
predicted_salary_poly = model.predict(age_range_wide_poly)

# Plotting the data points and polynomial regression curve
plt.scatter(age, salary, color='blue', label='Data Points')
plt.plot(age_range_wide, predicted_salary_poly, color='red', label='Polynomial Regression (Degree 2)')

# Labels and title
plt.xlabel('Age (Years)')
plt.ylabel('Salary (in Thousands)')
plt.title('Polynomial Regression: Age vs Salary')
plt.legend()
plt.grid(True)
plt.show()
```
